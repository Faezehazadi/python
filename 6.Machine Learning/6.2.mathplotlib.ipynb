```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
print(matplotlib.__version__)
```

Draw multiple points

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()

plt.plot(xpoints, ypoints, marker='o')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints,'o:m',ms=20 ,mec='m',mfc='0.8')
plt.show()
```

draw 2 lines

```python
points_1 = np.array([2, 6, 1, 10])
points_2 = np.array([8, 4, 10, 2])
plt.plot(points_1)
plt.plot(points_2)
plt.show()
```

Label for plots

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.show()
```

Sub plots

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y, 'r')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

#(rows, cols, index)
plt.subplot(2, 1, 2)
plt.plot(x,y, 'g')
plt.show()
```

Scatter

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color='purple')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color='orange')

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
x = np.array(["A", "B", "C", "D", "F"])
y = np.array([3, 8, 2, 5, 10])

plt.bar(x, y, color='m')
plt.show()

plt.barh(x, y, color='y')
plt.show()
```

```python
x = np.random.normal(170, 10, 250)

plt.hist(x, color='orange')
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0.1, 0, 0]

plt.pie(y, labels = mylabels,explode=myexplode)
plt.legend(title = "Four Fruits:", bbox_to_anchor=(1.05, 0.5), loc="center left")
plt.show()
```
