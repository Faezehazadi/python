```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
apples_width = np.random.normal(6, 1, n)
apples_length = np.random.normal(6, 1, n)
bananas_width = np.random.normal(4, 1, n)
bananas_length = np.random.normal(8, 1, n)
watermelons_width = np.random.normal(20, 1, n)
watermelons_length = np.random.normal(20, 1, n)
```

```python
width =  np.concatenate ((apples_width, bananas_width, watermelons_width))
length =  np.concatenate ((apples_length, bananas_length, watermelons_length))
x_train = np.array ([width, length]).T
x_train
```

```python
plt.scatter(apples_width, apples_length, c='red', marker='o', alpha=0.5)
plt.scatter(bananas_width, bananas_length, c='yellow', marker='o', alpha=0.5)
plt.scatter(watermelons_width, watermelons_length, c='green', marker='o', alpha=0.5)
```

```python
y_apples = np.zeros (n, dtype='int')
y_bananas = np.ones (n, dtype='int')
y_watermelons = np.ones (n, dtype='int')*2
y_train =  np.concatenate ((y_apples, y_bananas, y_watermelons))
y_train
```

```python
def Euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
new_fruit = np.array([20, 22])
distances= []
for x in x_train:
         d = Euclidean_distance (new_fruit, x)
         distances.append(d)

distances
```

```python
k = 5
nearest_neighbors = np.argsort(distances)[0:k]
nearest_neighbors
```

```python
y_train [nearest_neighbors]
```

```python
result = np.bincount(y_train[nearest_neighbors])

if np.argmax (result) == 0:
    print("")
elif np.argmax (result) == 1:
    print("")
else:
    print("")
```
